```python
import requests
from bs4 import BeautifulSoup
```

```python
r=requests.get("https://www.flipkart.com/search?sid=tyy%2C4io&otracker=CLP_Filters&p%5B%5D=facets.serviceability%5B%5D%3Dfalse&p%5B%5D=facets.brand%255B%255D%3DApple&sort=popularity")
c=r.content
```

```python
c
```

```python
soup=BeautifulSoup(c,"html.parser")
```

```python
all=soup.find_all("div",{"class":"_1UoZlX"})
```

```python
all
```

```python
l=[]
for item in all:
    d={}
    var_2=item.find("div",{"class":"_3wU53n"})
    d["Name"]=var_2.text
    var=item.find("div",{"class":"_1vC4OE _2rQ-NK"})
    var_3=var.text
    var_1=var_3.replace(',',"")
    d["Price"]=var_1.replace('',"") 
    l.append(d)
```

```python
l
```

```python
import pandas
df=pandas.DataFrame(l)
```

```python
df
```

```python
list_1=[]
for item in df["Name"]:
    try:
        d={}
        index_1=item.index('(')
        n_1=item[:index_1]
        print(n_1)
        d=item[:index_1]
    except:
        None
        d=item[:index_1]
    list_1.append(d)
    
list_2=[]
for item in df["Name"]:
    d={}
    try:
        index_2=item.index(',')
        n_2=item[index_2+1:].replace(')',"").replace(" ","")
        print(n_2)
        d=n_2
    except:
        None
    list_2.append(d)
    
```

```python
list_1
```

```python
list_2
```

```python
l2=[]
for item in df["Price"]:
    d={}
    d["Price"]=item
    l2.append(d)
```

```python
l2
```

```python
df_f1=pandas.DataFrame(list_1)
```

```python
df_f1
```

```python
df_f2=pandas.DataFrame(list_2)
```

```python
df_f2
```

```python
df_f3 =pandas.concat([df_f1, df_f2], axis=1, join='inner')
```

```python
df_f3
```

```python
df_f3.columns=['Name','Storage']
```

```python
df_f3
```

```python
df2=pandas.DataFrame(l2)
```

```python
df2
```

```python
df_concat = pandas.concat([df_f3, df2], axis=1, join='inner')
```

```python
df_concat
```

```python
df_concat['name_storage'] = df_concat['Name'] +'-' + df_concat['Storage'].map(str)
```
